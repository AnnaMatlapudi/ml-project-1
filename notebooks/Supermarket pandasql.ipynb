```python
import pandas as pd
from pandasql import sqldf
```

```python
df = pd.read_csv("Kaggle Supermarket Dataset.csv")
df.head()
```

```python
df = df.rename(columns={'Ship Mode':'Shipment_mode',
                   'Postal Code':'Postal_code',
                   'Sub-Category' :'Sub_category',
                  })
df.info()
```

```python
query = " SELECT Shipment_mode,Segment,Country, City, State, Region,Category, Sub_category FROM df "
df_orders = sqldf(query)
df_orders.head()
```

```python
query = " SELECT * \
        FROM df_orders \
        WHERE Region = 'South' "
df_south = sqldf(query)
df_south.head()
```

```python
query = " SELECT * \
        FROM df_south \
        WHERE Segment = 'Consumer' \
        AND City = 'Concord' "
df_south = sqldf(query)
df_south.head()
```

```python
query = " SELECT COUNT(*) \
        FROM df_south \
        WHERE Segment = 'Consumer' \
        AND City = 'Concord' "
df_south = sqldf(query)
df_south.head()
```

```python
query = " SELECT State,Category,Sub_category \
        FROM df_orders \
        WHERE State = 'Florida' \
        AND Category = 'Furniture' \
        AND Sub_category = 'Tables' "
df_florida = sqldf(query)
df_florida.head()
```

```python
query =" SELECT COUNT(*) FROM df_florida "
df_florida_count = sqldf(query)
df_florida_count.head()
```

```python
query = " SELECT * \
        FROM df \
        WHERE City = 'San Francisco' \
        AND Quantity < 2 \
        AND Shipment_mode IN ('First Class', 'Second Class') "
df_1 = sqldf(query)
df_1.head()
```

```python
query = " SELECT *,\
          COUNT (Shipment_mode) AS ShippingMode\
          FROM df_1 \
          GROUP BY Segment\
          ORDER BY City "
df_grp = sqldf(query)
df_grp
```

```python
query = " SELECT Segment,Shipment_mode,City,\
          COUNT (Shipment_mode) AS ShippingMode\
          FROM df_1 \
          GROUP BY Segment\
          ORDER BY City "
df_grp = sqldf(query)
df_grp
```

```python
query = " SELECT City, Shipment_mode,Sales,Quantity,Discount,Profit \
        FROM df \
        GROUP BY Shipment_mode\
        Order By City "

df_sales = sqldf(query)
df_sales.head()
```

```python
query = "SELECT Sales,Quantity,Discount,Profit,Segment,Category, \
        COUNT(Quantity) AS Orderqty,MAX(Sales),MAX(Discount) \
        FROM df \
        GROUP BY Segment\
        Order By Sales "

df_salesqty = sqldf(query)
df_salesqty.head(10)
```

```python
query = " SELECT Sales,Quantity,Discount,Profit, \
        COUNT (Profit) \
        FROM df_sales \
        GROUP BY City "

df_sales1 = sqldf(query)
df_sales1.head(10)
```
